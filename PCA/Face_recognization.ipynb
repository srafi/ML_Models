```python
#creating training matrix, each detected face is covered as vector
# if face image is 100*100, D = 100*100 =10000
def load_images(dataset_path):
    images =[]
    labels =[]
    label_map ={}
    label_id =0
    for filename in os.listdir(dataset_path):
        filepath = os.path.join(dataset_path, filename)
        print("filepath:", filepath)
        print("filename:", filename)
        if os.path.isfile(filepath):
            print("img_path:", filepath)
            img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, IMAGE_SIZE)
            images.append(img.flatten())
            labels.append(label_id)
        label_id +=1
    return np.array(images), np.array(labels), label_map
    # for person in os.listdir(dataset_path):
    #     person_path = os.path.join(dataset_path, person)
    #     if not os.path.isdir(person_path):
    #         continue
    #     label_map[label_id] = person
    #     for image_name in os.listdir(person_path):
            #img_path = os.path.join(filepath, filename)
           
X, Y, label_map = load_images(DATASET_PATH)
X =X.astype(np.float32)
print("Training Images:", X.shape)
```

```python
#Mean Faces
mean_faces = np.mean(X, axis=0)
#Mean centering
A = X- mean_faces
```

```python
# covariance matrix size
L = np.dot(A, A.T)
```

```python
# eigen faces are principle components of face variance 
idx = np.argsort(-eigenvalues)
eigenvectors = eigenvectors[:,idx]
eigenfaces = np.dot(A.T, eigenvectors[:, :NUM_EIGENFACES])
#eigenfaces = eigenfaces[:, :NUM_EIGENFACES]
```

```python
def recognize(image_path, threshold=3000):
    face = detect_face(image_path)
    print(face)
    print(face.shape)
    if face is None:
        return "No face detected"
    face = face.flatten().astype(np.float32)
    print(face.shape)
    print(mean_faces.shape)
    face_centered = face- mean_faces
    test_features = np.dot(face_centered, eigenfaces)
    print("train features", train_features)
    print("test features", test_features)
    distances = np.linalg.norm(train_features - test_features, axis=1)
    print("distances:",distances)
    min_index = np.argmin(distances)
    print(min_index)
    if distances[min_index] > threshold:
        return "Unknown"
    return distances[min_index]
```

# ========PCA Face Recognization========

```python
import cv2
import numpy as np
import os
from sklearn.decomposition import PCA
from PIL import Image
```

```python
from sklearn import decomposition
from sklearn.preprocessing import LabelEncoder
```

```python
IMAGE_SIZE = (100, 100)
N_COMPONENTS = 10
directory_path = 'face_recog/'
```

```python
smallest_image = None
smallest_size = None
smallest_dims = None
for person in os.listdir(directory_path):
    person_path = os.path.join(directory_path, person)
    if not os.path.isdir(person_path):
        continue
    for filename in os.listdir(person_path):
        filepath = os.path.join(person_path, filename)
        if os.path.isfile(filepath):
            img = Image.open(filepath)
            width, height = img.size
            area = width * height
            if smallest_size is None or area < smallest_size:
                smallest_size = area
                smallest_image = filepath
                smallest_dims = (width, height)
print("Smallest image:", smallest_image)
print("Size (width, height):", smallest_dims)
print("Total pixels:", smallest_size)
```

```python
for person in os.listdir(directory_path):
        person_path = os.path.join(directory_path, person)
        if not os.path.isdir(person_path):
            continue
        for filename in os.listdir(person_path):
            filepath = os.path.join(person_path, filename)
            if os.path.isfile(filepath):
                filename = os.path.basename(filepath)
                img = Image.open(filepath)
                img_array = np.array(img)
                #output_img = np.zeros(img_array.shape, dtype=np.int64)
                output_array = np.zeros_like(img_array)
                height, width, channels = img_array.shape
                for i in range(height):
                    for j in range(width):
                        avg_value = int(np.mean(img_array[i, j]))
                        for c in range(channels):
                            output_array[i, j, c] = avg_value
                output_img = Image.fromarray(output_array)
                resized_img = output_img.resize(smallest_dims, Image.LANCZOS)
                resized_img.save(person_path+"/"+filename)
```

```python
def load_faces(dataset_path):
    faces = []
    labels = []

    for person in os.listdir(dataset_path):
        person_path = os.path.join(dataset_path, person)
        if not os.path.isdir(person_path):
            continue
        for img_name in os.listdir(person_path):
            img_path = os.path.join(person_path, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue
            img = cv2.resize(img, IMAGE_SIZE)
            faces.append(img.flatten())
            labels.append(person)
    return np.array(faces), np.array(labels)
```

```python
### print("[INFO] Loading training data...")
X, y = load_faces(directory_path)

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("[INFO] Training PCA model...")
pca = PCA(n_components=N_COMPONENTS, whiten=True)
X_pca = pca.fit_transform(X)
```

```python
# -----------------------------
# COMPUTE THRESHOLD
# -----------------------------
print("[INFO] Computing threshold...")
intra_distances = []

for i in range(len(X_pca)):
    for j in range(len(X_pca)):
        if y_encoded[i] == y_encoded[j] and i != j:
            d = np.linalg.norm(X_pca[i] - X_pca[j])
            intra_distances.append(d)

intra_distances = np.array(intra_distances)
THRESHOLD = np.mean(intra_distances) + 2 * np.std(intra_distances)

print(f"[INFO] Threshold value: {THRESHOLD:.2f}")
```

```python
# -----------------------------
# FACE RECOGNITION FUNCTION
# -----------------------------
def recognize_face(test_image_path):
    img = cv2.imread(test_image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError("Test image not found")

    img = cv2.resize(img, IMAGE_SIZE)
    test_vector = img.flatten()

    # Project test image into PCA space
    test_pca = pca.transform([test_vector])

    # Compute Euclidean distances
    distances = np.linalg.norm(X_pca - test_pca, axis=1)

    min_index = np.argmin(distances)
    min_distance = distances[min_index]
    predicted_label = y_encoded[min_index]

    name = label_encoder.inverse_transform([predicted_label])[0]

    return name, min_distance
```

```python
# -----------------------------
# TEST IMAGE
# -----------------------------
print("[INFO] Testing image...")
img = Image.open("face_recog/sneha.JPG")
img_array = np.array(img)
output_array = np.zeros_like(img_array)
height, width, channels = img_array.shape
for i in range(height):
    for j in range(width):
        avg_value = int(np.mean(img_array[i, j]))
        for c in range(channels):
            output_array[i, j, c] = avg_value
output_img = Image.fromarray(output_array)
resized_img = output_img.resize(smallest_dims, Image.LANCZOS)
resized_img.save("face_recog/sneha_b.JPG")
test_image_path = "face_recog/sneha_b.JPG"

name, distance = recognize_face(test_image_path)

print(f"Predicted: {name}")
print(f"Distance : {distance:.2f}")

if distance < THRESHOLD:
    print(" Face Recognized")
else:
    print(" Unknown Face")
```

```python
# FIX #2: Reconstruction Error
def reconstruction_error(test_vector):
    projected = pca.transform([test_vector])
    reconstructed = pca.inverse_transform(projected)
    error = np.linalg.norm(test_vector - reconstructed[0])
    return error
```

```python
recon_errors = []
for x in X:
    err = reconstruction_error(x)
    recon_errors.append(err)
RECON_THRESHOLD = np.mean(recon_errors) + 2 * np.std(recon_errors)
```
